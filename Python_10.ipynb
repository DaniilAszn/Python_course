```python
class MealyError(Exception):
    def __init__(self, message):
        super().__init__(message)


class Mealy:
    def __init__(self):
        self.state = 'A'

    def code(self):
        if self.state == 'A':
            self.state = 'B'
            return 0
        if self.state == 'B':
            self.state = 'C'
            return 2
        if self.state == 'F':
            self.state = 'C'
            return 8
        else:
            raise MealyError('code')

    def pluck(self):
        if self.state == 'B':
            self.state = 'B'
            return 3
        if self.state == 'D':
            self.state = 'E'
            return 5
        if self.state == 'F':
            self.state = 'A'
            return 9
        else:
            raise MealyError('pluck')

    def paint(self):
        if self.state == 'C':
            self.state = 'D'
            return 4
        if self.state == 'E':
            self.state = 'F'
            return 6
        if self.state == 'F':
            self.state = 'G'
            return 7
        if self.state == 'A':
            self.state = 'E'
            return 1
        else:
            raise MealyError('paint')


def test():
    o = main()
    o.paint()  # 1
    o.paint()  # 6
    o.code()  # 8
    o.paint()  # 4
    try:
        o.paint()  # MealyError
    except MealyError:
        pass
    o.pluck()  # 5
    o.paint()  # 6
    o.pluck()  # 9
    o.code()  # 0
    try:
        o.paint()  # MealyError
    except MealyError:
        pass
    o.pluck()  # 3
    o.code()  # 2

    o = main()
    o.code()  # 0
    o.pluck()  # 3
    o.code()  # 2
    o.paint()  # 4
    o.pluck()  # 5
    o.paint()  # 6
    o.code()  # 8
    o.paint()  # 4
    o.pluck()  # 5
    o.paint()  # 6
    o.pluck()  # 9
    o.paint()  # 1
    o.paint()  # 6
    o.paint()  # 7

    try:
        o.pluck()  # MealyError
    except MealyError:
        pass
    try:
        o.code()  # MealyError
    except MealyError:
        pass


def main():
    return Mealy()
```
