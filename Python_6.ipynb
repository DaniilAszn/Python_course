```python
s = (
    {2014, 2018, 'XTEND', 1978},
    {2014, 2018, 'XTEND', 1966},
    {2014, 2018, 'QML'},
    {2014, 1969, 1990, 1978},
    {2014, 1969, 1990, 1966},
    {2014, 1969, 2008, 'XTEND'},
    {2014, 1969, 2008, 'QML'},
    {2014, 1969, 1982},
    {2014, 1981},
    {2006, 2018, 1990, 'XTEND'},
    {2006, 2018, 1990, 'QML'},
    {2006, 2018, 2008},
    {2006, 2018, 1982},
    {2006, 1969},
    {2006, 1981},
    {1962}
)


def main(n):
    S = set(n)
    for i, v in enumerate(s):
        if not (len(v - S)):
            return i
```

```python
def zero(items, left, right):
    if items[0] == 'XTEND':
        return right
    if items[0] == 'QML':
        return left


def one(items, left, middle, right):
    if items[1] == 1990:
        return right
    if items[1] == 2008:
        return middle
    if items[1] == 1982:
        return left


def two(items, left, middle, right):
    if items[2] == 2018:
        return right
    if items[2] == 1969:
        return middle
    if items[2] == 1981:
        return left


def three(items, left, middle, right):
    if items[3] == 2014:
        return right
    if items[3] == 2006:
        return middle
    if items[3] == 1962:
        return left


def four(items, left, right):
    if items[4] == 1978:
        return right
    if items[4] == 1966:
        return left


def main(items):
    result = three(
        items,
        15,
        two(
            items,
            14,
            13,
            one(
                items,
                12,
                11,
                zero(
                    items,
                    10,
                    9
                )
            )
        ),
        two(
            items,
            8,
            one(
                items,
                7,
                zero(
                    items,
                    6,
                    5
                ),
                four(
                    items,
                    4,
                    3
                )
            ),
            zero(
                items,
                2,
                four(
                    items,
                    1,
                    0
                )
            )
        )
    )
    return result


print(main(['QML', 1990, 1981, 2006, 1978]))
```
