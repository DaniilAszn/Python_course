```python
def main(P):
    G = {4 * p - p ** 5 for p in P if ((p < -10) or (p > 8))}
    D = {abs(p) - 10 * p for p in P if ((p > -10) and (p < -1))}
    cap_D = len(D)
    prod_G = 1
    for p in G:
        prod_G *= p
    a = cap_D - prod_G
    return a


print(main([4.73, 0.62, 6.07, 5.34, 8.14]))
```
