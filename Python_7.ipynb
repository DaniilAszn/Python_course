```python
def main(data):
    binary_data = int(data, 16)
    result = []
    result.append(('C1', hex(0b111 & binary_data)))
    result.append(('C2', hex(0b1111 & (binary_data >> 3))))
    result.append(('C3', hex(0b11 & (binary_data >> 7))))
    result.append(('C4', hex(0b111111111 & (binary_data >> 9))))

    return result


print(main('0x3d2e9'))



# str = '0x12531'
# s = int(str, 16)
# c1 = 0b111 & s
# print(f'c1: {hex(c1)}')
#
# c2 = 0b1111 & (s >> 3)
# print(f'c2: {hex(c2)}')
#
# c3 = 0b11 & (s >> 7)
# print(f'c3: {hex(c3)}')
#
# c4 = 0b111111111 & (s >> 9)
# print(f'c4: {hex(c4)}')
```
