```python
import math as m


def main(z, x):
    n = len(z)
    s = 0
    for i in range(1, n + 1):
        s += 80 * (3 * (z[n + 1 - i - 1]) ** 3 +
                   44 * (x[m.ceil(i / 3) - 1]) ** 2) ** 4
    return 32 * s
```

```python
from math import ceil


def main(z, x):
    n = len(z)
    s = 0
    for i in range(1, n + 1):
        s += 80 * (3 * (z[n + 1 - i - 1]) ** 3 +
                   44 * (x[ceil(i / 3) - 1]) ** 2) ** 4
    return 32 * s


print(main([-0.23, -0.53, 0.71, -0.69, 0.65],
[-0.39, 0.25, 0.86, -0.85, -0.22]))
```
