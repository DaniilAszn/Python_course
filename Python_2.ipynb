```python
import math


def main(z):
    if (z < 152):
        return z ** 3 / 46
    elif (152 <= z) and (z < 200):
        return 1 + 96 * (z / 9 - z ** 2 - 1) ** 7
    elif (200 <= z) and (z < 263):
        return (math.floor(z)) ** 6 - 22 * z - (math.atan(z)) ** 3
    elif (263 <= z) and (z < 350):
        return 67 * (51 * z ** 2 - z - 1) ** 3
    elif (z >= 350):
        return (8 * z ** 2 + 1) ** 6
```
