```python
def main(n):
    if (n == 0):
        return 0.39
    if (n == 1):
        return -0.94
    if (n >= 2):
        return main(n-1) - 1 - (main(n-2))**3
```

```python
def main(n):
    cache = [0.39, -0.94]
    if n < len(cache):
        return cache[n]
    else:
        result = main(n - 1) - 1 - main(n - 2) ** 3
        cache.append(result)
        return result
```

```python
def main(n, a=0.39, b=-0.94):
    return (
        a if n == 0 else
        b if n == 1 else
        main(n - 1, b, b - 1 - (a ** 3))
    )
```

```python
def main(n):
    a, b = 0.39, -0.94
    for _ in range(n):
        a, b = b, b - 1 - (a ** 3)
    return a


print(main(8))
```
