```python
import math as m


def main(z, x):
    result = m.sqrt(51 * (9 * x - 42 * z ** 3 - 1) ** 6) - \
             (69 * (m.sin(z)) ** 7 - 63 * (m.cos(1 - 45 * x ** 3)) ** 2)
    return result
```

```python
from math import sqrt, sin, cos


def main(z, x):
    a = 51
    b = (9 * x - 42 * z ** 3 - 1) ** 6
    c = sqrt(a * b)
    d = 69
    e = (sin(z)) ** 7
    f = 63
    g = (cos(1 - 45 * x ** 3)) ** 2
    h = d * e - f * g
    return c - h
```
