```python
import math as m


def main(b, a, z, n):
    s1 = 0
    for c in range(1, a + 1):
        for i in range(1, b + 1):
            s1 += (m.floor(c)) ** 6 - 22 * z - (m.atan(i)) ** 3
    s2 = 0
    for j in range(1, n + 1):
        for i in range(1, b + 1):
            s2 += 8 * j ** 6 + z ** 3 + ((i ** 2) / 56 - 1) ** 3
    s = s1 - s2
    return s
```

```python
import math


def main(b, a, z, n):
    s = 0
    for c in range(1, a + 1):
        for i in range(1, b + 1):
            s += (math.floor(c)) ** 6 - 22 * z - (math.atan(i)) ** 3
    for j in range(1, n + 1):
        for i in range(1, b + 1):
            s -= 8 * j ** 6 + z ** 3 + ((i ** 2) / 56 - 1) ** 3
    return s


print(main(8, 7, -0.64, 6))
```
