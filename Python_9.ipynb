```python
'''
ДЛЯ РЕШЕНИЯ ЛУЧШЕ ЮЗАТЬ
import copy

def main(table_old):
    table_new = copy.deepcopy(table_old)
    return table_new
'''

def transform_table(table):
    # удаление дубликатов среди первых двух столбцов
    for i in range(len(table)):
        if table[i][0] == table[i][1]:
            table[i].remove(table[i][1])
    # форматирование по примеру
    for i in range(len(table)):
        table[i][0] = table[i][0].replace('@', '[at]')
        table[i][1] = table[i][1].replace('.', '-')
        table[i][2] = table[i][2].replace('-', ' ', 1)
        table[i][3] = (str(round(float(table[i][3]), 3)) + '00')[:5]
    # сортировка строк по третьему столбцу
    a = []
    for i in range(len(table)):
        a.append(int(table[i][2][:4]))
    for i in range(len(table)):
        for j in range(i + 1, len(table)):
            if (a[i] > a[j]):
                t = table[i]
                table[i] = table[j]
                table[j] = t
                ta = a[i]
                a[i] = a[j]
                a[j] = ta
    # транспонирование таблицы
    trans = []
    for i in range(len(table[0])):
        ci = []
        for j in range(len(table)):
            a = table[j][i]
            ci.append(a)
        trans.append(ci)
    table = trans

    return table


def main(table):
    return transform_table(table)


table = [
    ['bodudli85@yahoo.com', 'bodudli85@yahoo.com', '21.09.00', '528-386-0890', '0.8745'],
    ['rotuk85@gmail.com', 'rotuk85@gmail.com', '14.04.99', '023-067-6973', '0.6643'],
    ['dirov52@yandex.ru', 'dirov52@yandex.ru', '03.05.99', '747-176-1435', '0.2059'],
    ['tuzuzman42@yahoo.com', 'tuzuzman42@yahoo.com', '01.08.99', '544-187-1336', '0.4874']
]
print(main(table))
```
